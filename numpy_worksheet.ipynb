### Exercise 15 — Vectorize a Function

**Question:** Convert the scalar function `maxx` below to work on two arrays.


```python
# Input — do not modify
def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y

a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
```

```python
# Write your code here:
maxx_vectorized = np.vectorize(maxx)
answer = maxx_vectorized(a, b)
print(answer)
```

### Exercise 28 — Mean, Median, Std

**Question:** Find the mean, median, and standard deviation of the sepal length (1st column) of `iris_2d`.

*# Make sure you have run Exercise 27 first.*


```python
sepal_length = iris_2d[:, 0]
mean_sepal_length = np.mean(sepal_length)
median_sepal_length = np.median(sepal_length)
std_sepal_length = np.std(sepal_length)

print(f"Mean of sepal length: {mean_sepal_length:.2f}")
print(f"Median of sepal length: {median_sepal_length:.2f}")
print(f"Standard deviation of sepal length: {std_sepal_length:.2f}")
```

### Exercise 29 — Normalize Array

**Question:** Create a normalized form of iris's sepal length whose values range exactly between 0 and 1 (minimum → 0, maximum → 1).

*# Make sure you have run Exercise 27 first.*


```python
sepal_length = iris_2d[:, 0]
min_sepal_length = np.min(sepal_length)
max_sepal_length = np.max(sepal_length)
normalized_sepal_length = (sepal_length - min_sepal_length) / (max_sepal_length - min_sepal_length)
print(normalized_sepal_length)
```
